```python
def I(t):
    if t == 0:
        it = I_0
    else:
        it = I(t-1) + beta*S(t-1)*I(t-1)/N - mu*I(t-1)
    return it
```

```python
def S(t):
    if t == 0: 
        st = S_0
    else: 
        st = S(t-1) - beta*S(t-1)*I(t-1)/N
    return st
```

```python
def R(t):
    if t == 0: 
        rt = R_0
    else: 
        rt = R(t-1) + mu*I(t-1)
    return rt
```

```python
def SIR(t):
    s = []
    i = []
    r = []
    for m in range(t): 
        s.append(S(m))
        i.append(I(m))
        r.append(R(m))
    return s,i,r
```

```python
def plotter(t):
    x = range(t)
    y = SIR(t)
    plt.plot(x, y[0], label = 'S')
    plt.plot(x,y[1], label = 'I')
    plt.plot(x,y[2], label = 'R')
    plt.xlabel('time')
    plt.ylabel('nodes')
    plt.legend()
    plt.show()
```

```python
def I_beta(t, beta):
    if t == 0:
        it = I_0
    else:
        it = I(t-1) + beta*S(t-1)*I(t-1)/N - mu*I(t-1)
    return it
```

```python
def S_beta(t,beta):
    if t == 0: 
        st = S_0
    else: 
        st = S(t-1) - beta*S(t-1)*I(t-1)/N
    return st
```

```python
def SIR_beta(t,beta):
    s = []
    i = []
    r = []
    for m in range(t): 
        s.append(I_beta(m,beta[0]))
        i.append(I_beta(m,beta[1]))
        r.append(I_beta(m,beta[2]))
    return s,i,r
```

```python
def plotter_beta(t,beta):
    x = range(t)
    y = SIR_beta(t,beta)
    plt.plot(x, y[0], label = 'beta 1')
    plt.plot(x,y[1], label = 'beta 2')
    plt.plot(x,y[2], label = 'beta 3')
    plt.xlabel('time')
    plt.ylabel('nodes')
    plt.legend()
    plt.show()
```
